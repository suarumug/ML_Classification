```python
import pandas as pd
```

```python
dataset= pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset=dataset.drop("User ID",axis=1)
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['Age', 'EstimatedSalary','Gender_Male']]
dependent=dataset[['Purchased']]
```

```python
# training and test set split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
# model creation -GaussianNB
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
y_predict=classifier.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)

from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_predict)
print(clf_report)
print(cm)
```

```python
# model creation -MultinomialNB
from sklearn.naive_bayes import MultinomialNB 
classifier=MultinomialNB()
classifier.fit(x_train,y_train)
y_predict=classifier.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)

from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_predict)
print(clf_report)
print(cm)
```

```python
# model creation -BernoulliNB
from sklearn.naive_bayes import BernoulliNB 
classifier=BernoulliNB()
classifier.fit(x_train,y_train)
y_predict=classifier.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)

from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_predict,zero_division=0)
print(clf_report)
print(cm)
```

```python
# model creation -CategoricalNB
from sklearn.naive_bayes import CategoricalNB 
classifier=CategoricalNB()
classifier.fit(x_train,y_train)
y_predict=classifier.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)

from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_predict,zero_division=0)
print(clf_report)
print(cm)
```

```python
# model creation -ComplementNB
from sklearn.naive_bayes import ComplementNB 
classifier=ComplementNB()
classifier.fit(x_train,y_train)
y_predict=classifier.predict(x_test)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)

from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_predict,zero_division=0)
print(clf_report)
print(cm)
```
